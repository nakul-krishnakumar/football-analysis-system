```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "../output_videos/player_1.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image[: 1, : 5, :] # [row, column, rgb(r, g, b)]
# image[0][0] -> rgb of cell in row 0 and column 0
```

```python
plt.imshow(image)
plt.show()
```

# Taking Top Half of The Image

```python
# Taking only top half as to analyze only the jersey color
top_half_img = image[:len(image)//2] 

plt.imshow(top_half_img)
plt.show()
```

# Cluster the image into two clusters

```python
# Reshape the image into 2D array
image_2d = top_half_img.reshape(-1,3)

image_2d # [row * column, rgb]
```

```python
# Perform KMeans clustering with 2 clusters
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(image_2d)

# Get the cluster labels
labels = kmeans.labels_

# Reshape labels into original image shape
clustered_img = labels.reshape(top_half_img.shape[0], top_half_img.shape[1])

# Display the results
plt.imshow(clustered_img)
plt.show()
```

```python
corner_clusters = [clustered_img[0, 0], clustered_img[0, -1], clustered_img[-1, 0], clustered_img[-1, -1]] # list of 1s and 0s
# 1 -> then the corner point is not a jersey
# 0 -> then the corner point is a jersey

non_player_cluster = max(set(corner_clusters), key=corner_clusters.count)

print(non_player_cluster) # if 1, then player cluster has label
```

```python
player_cluster = 1 - non_player_cluster
print(player_cluster)
```

# Verify if player cluster label is correct

```python
rgb = kmeans.cluster_centers_[player_cluster]
rgb
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4))

# Displaying color of the player cluster
square = plt.Rectangle((0,0), 1, 1, color=rgb/255)
ax[0].add_patch(square)
ax[0].set_xlabel("Player Cluster Color")

# Displaying top half of the player image
ax[1].imshow(top_half_img)
ax[1].set_xlabel("Top Half of the Player")
plt.show()
```
